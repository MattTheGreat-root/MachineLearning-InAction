```python
import cv2
img = cv2.imread("./Images/person.jpg")
```

```python
import matplotlib.pyplot as plt
plt.imshow(img[... , ::-1])
```

```python
img.shape
```

```python
import numpy as np

cropped = img[10:140, 210:340]
circle = np.zeros((130,130,3), np.uint8)
h, w = cropped.shape[:2]

circle = np.zeros((h, w), np.uint8)
cv2.circle(circle, (w//2, h//2), min(h,w)//2, 255, -1)
output_image = cv2.bitwise_and(cropped,cropped, mask=circle)
plt.imshow(output_image[..., ::-1])
```
