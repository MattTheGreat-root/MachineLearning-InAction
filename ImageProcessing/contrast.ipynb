```python
import cv2
img = cv2.imread("./images/dark.png")
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

Linear Correction

```python
new_image = cv2.convertScaleAbs(img, alpha=2, beta=10)
plt.imshow(new_image)
```

Gamma correction

```python
import numpy as np

gamma = 0.4

lookUpTable = np.empty((1,256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    
output_gammacorrected = cv2.LUT(img, lookUpTable)

plt.imshow(output_gammacorrected)
```
